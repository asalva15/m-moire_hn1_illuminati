```python
import pytesseract
from PIL import Image

def perform_ocr(image_path):
    # Open the image file
    image = Image.open(image_path)

    # Use pytesseract to do OCR on the image and get the OCR result with confidence scores
    result = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

    # Extract confidence scores and filter out empty strings
    confidences = [int(conf) for conf in result['conf'] if conf]

    # Calculate the average confidence score
    if confidences:
        average_confidence = sum(confidences) / len(confidences)
    else:
        average_confidence = 0

    # Normalize the average confidence score to a percentage scale
    certainty_percentage = (average_confidence / 100) * 100

    # Get the recognized text
    recognized_text = pytesseract.image_to_string(image)

    return recognized_text, certainty_percentage

# Example usage
image_path = 'photo_essai_tesseract.jpg'
recognized_text, certainty_percentage = perform_ocr(image_path)

print("Recognized text:", recognized_text)
print("Certainty percentage:", certainty_percentage)
```

```python
from spellchecker import SpellChecker

def spell_check_text(input_file, output_file, language='en'):
    # Create an instance of SpellChecker
    spell = SpellChecker(language=language)

    # Read input file
    with open(input_file, 'r', encoding='utf-8') as file:
        text = file.read()

    # Split the text into words
    words = text.split()

    # Find misspelled words
    misspelled = spell.unknown(words)

    # Correct misspelled words
    for word in misspelled:
        corrected_word = spell.correction(word)
        if corrected_word is not None:
            text = text.replace(word, corrected_word)

    # Write corrected text to output file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(text)

    print("Spell-checking completed. Corrected text saved to:", output_file)

# Example usage
input_file = 'essai1.txt'
output_file = 'essai1_corrected_output.txt'
spell_check_text(input_file, output_file)
```
